```python
def missing_values(immo_data,norows): 
    total = immo_data.isnull().sum().sort_values(ascending=False)
    percent = ((immo_data.isnull().sum().sort_values(ascending=False)/immo_data.shape[0])*100).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return(missing_data.head(norows))
```

```python
# az immo_data estefade mikonim va tedad satrha 20 ast
missing_values(immo_data,20) 
```

```python
#har sotuni ke null bish az 30% ast drop mikonam
missing_data = missing_values(immo_data,20)
immo_data = immo_data.drop((missing_data[missing_data['Percent'] > 30]).index,1)
```

```python
#keep total rent
immo_data.dropna(subset=['totalRent'],inplace=True)
```
